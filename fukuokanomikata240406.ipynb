```python
import pandas as pd

# 山手線の駅名リスト
stations = [
    "東京", "有楽町", "新橋", "浜松町", "田町", "品川", "高輪ゲートウェイ","大崎", "五反田", "目黒", "恵比寿",
    "渋谷", "原宿", "代々木", "新宿", "新大久保", "高田馬場", "目白", "池袋", "大塚", "巣鴨",
    "駒込", "田端", "西日暮里", "日暮里", "鶯谷", "上野", "御徒町", "秋葉原", "神田"
]

# 空のDataFrameを用意
combined_df = pd.DataFrame()

# 各駅のCSVファイルを読み込み、駅名列を追加して結合
for station in stations:
    filename = f'{station}_data.csv'
    df = pd.read_csv(filename)
    df['Station'] = station  # 駅名列を追加
    combined_df = pd.concat([combined_df, df], ignore_index=True)

# 結合したデータを新しいCSVファイルに保存
combined_df.to_csv('yamanote_data.csv', index=False, encoding='utf-8')

print("全駅のデータを結合したCSVファイルが保存されました: yamanote_data.csv")
```

```python
# "0120"で始まる番号を含む行を除外
# NaNを含む場合でもエラーにならないようにするために、fillnaを使用してNaNを空文字列に置き換える
df_filtered = df[df['Tel1'].fillna('').str.startswith('0120') == False]

# Tel1を基にグループ化し、各グループ内のレコード数をカウント
grouped = df_filtered.groupby(['Tel1']).size().reset_index(name='counts')

# レコード数が1より大きいグループのTel1を抽出
duplicates = grouped[grouped['counts'] > 1][['Tel1']]

# 重複データの抽出
duplicated_data = pd.merge(df_filtered, duplicates, on=['Tel1'])

# 結果の表示（最初の20行を表示）
print(duplicated_data.head(20))

```

```python
## 各都市名ごとの合計数を計算
totals = grouped_df[locations].sum()

# 合計数を降順にソートしてランキングを作成
ranking = totals.sort_values(ascending=False)

# ランキング結果をDataFrameに変換
ranking_df = ranking.reset_index()
ranking_df.columns = ['Location', 'Total']

# 結果をCSVファイルに保存
ranking_df.to_csv('location_totals_ranking.csv', index=False, encoding='utf-8')

print("都市名ごとの合計数ランキングがCSVファイルに保存されました: location_totals_ranking.csv")
```
